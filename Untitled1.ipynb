```r
set.seed(1)
par(bg = "white", col.axis = "white", col.lab = "white", col.main = "white", 
    col.sub = "white", xaxt = "n", yaxt = "n", bty = "n") # 흰 캔버스 생성
plot(x = c(0, 10), y = c(0, 10), type = "n") # 범위만 설정
for (i in 0:10) {
  lines(x = c(0, 10), y = rep(i, 2), lty = 1)
} # 수평선 그리기


# 바늘 떨어뜨리기

n=10000 # 시행횟수
hits = 0 # 빨간선 count 초기화
for (i in 1:n) {
  x1 <- runif(1, 0, 10) # 바늘의 중간 x좌표
  y1 <- runif(1, 0, 10) # 바늘의 중간 y좌표
  angle <- runif(1, 0, 2*pi) # 바늘 기울기 결정
  
  x2 <- x1 + 0.3 * cos(angle + pi) # 바늘의 한 쪽 끝 x좌표 결정
  y2 <- y1 + 0.3 * sin(angle + pi) # 바늘의 한 쪽 끝 y좌표 결정
  
  x3 <- x1 + 0.3 * cos(angle) # 바늘의 한 쪽 끝 x좌표 결정
  y3 <- y1 + 0.3 * sin(angle) # 바늘의 한 쪽 끝 y좌표 결정
  
  # 바늘이 수평선 사이에 위치하는 가?
  # 바늘의 양 끝 y좌표를 각각 내림하고 올림했을 때 동일하면 수평선 사이에 위치한 것
  # 이때의 col = "blue"
  if (floor(y2) == floor(y3) & ceiling(y2) == ceiling(y3)) {
    col <- "blue"
  } else {
    col <- "red"
    hits <- hits + 1 # 수평선에 닿았으니 카운트 +1
  }
  
  lines(x = c(x2, x3), y = c(y2, y3), col = col)
}

# pi 추정하기
pi_estimate <- 2 * 0.6 * n / hits
pi_estimate

mtext(side = 3, text = paste("pi estimate =", round(pi_estimate, 4), "\n n =", n), 
      font=2,line = 1)
```
