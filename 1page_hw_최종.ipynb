```r
score = read.csv("score.csv")
```

```r
score[score$practical>30,"class"]="A"
score[score$practical>20 & score$practical<31 ,"class"]="B"
score[score$practical>10 & score$practical<21 ,"class"]="C"
score[score$practical<11 ,"class"]="D"
```

```r
score[score$written>30,"class1"]="A"
score[score$written>20 & score$written<31 ,"class1"]="B"
score[score$written>10 & score$written<21 ,"class1"]="C"
score[score$written<11 ,"class1"]="D"
```

```r
classs = function(x){
    for(i in 1:37){
        if(x[i]>30){
            score$class[i]="A"
            }else if(x[i]>20){
            score$class[i]="B"
            }else if (x[i]>10){
            score$class[i]="C"
            }else{
            score$class[i]="D"}}
}
```
