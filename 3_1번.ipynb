# 2번

```r
mean_x = mean(x)
mean_y = mean(y)
```

```r
sxy = sum((x-mean_x)*(y-mean_y))
sxx = sum((x-mean_x)^2)
```

```r
b_1 = sxy/sxx
b_0 = mean_y - b1*mean_x
```

### $\therefore$ 최소제곱법에 의한 회귀직선은 다음과 같다

$y = b_0 + b_1 x$

```r
b_0
b_1
```

### 맞는지 확인해보자

```r
lm(y~x)
```

# 3번

- 추정된 회귀직선은 다음과 같다.

```r
y_hat = 29.11 + 13.64*x
```

```r
s_yx_2 = sum((y-y_hat)^2)/(14-2)
```

- 회귀로부터의 평균제곱편차

```r
s_yx_2
```

- 회귀에서의 표준편차 

```r
s_yx = sqrt(s_yx_2)
s_yx
```

- 검산해보자

```r
summary(lm(y~x))
```

### Residual standard error: 29.11 on 12 degrees of freedom
