## `1-(11)`

```r
yhat = fitted(full_model)
resd = resid(full_model)

plot(resd ~ yhat, pch = 16) # 잔차 산점도
abline(h=0,col="red",lwd=3) # y=0 수평선
```

- 0을 중심으로 대칭이며 일부 특이점을 제외하면 등분산성이라 할 수 있겠다.

## `1-(12)`

- $H_0$ : 등분산
- $H_1$ : 이분산

```r
library(lmtest)
bptest(full_model) # 잔차에 대한 등분산성 검정 수행
```

- p-value : 0.99이며 주어진 유의수준보다 크기 때문에 등분산성을 만족한다고 할 수 있다. 

---

## `1-(13)`

```r
par(mfrow=c(1,2))
# QQ plot 
qqnorm(resd, pch=16)
qqline(resd, col=2)

hist(resd) # 잔차 히스토그램
par(mfrow=c(1,1))
```

```r
# Shapiro-Wilk Test
# H0 : normal distribution, H1 : not H0
shapiro.test(resd) # 잔차에 대한 정규성 검정 수행
```

- p-value가 주어진 유의수준보다 작기 때문에 정규분포를 따른다고 볼 수 없다. 
- 왼쪽으로 꼬리가 긴 형태인 것으로 보인다. 
- 잔차가 특히 작은 값들이 존재하기 때문에, 마치 등분산성을 만족하지 않는 것으로 보일 수도 있다.

## `1-(14)`

```r
library(lmtest)
dwtest(full_model, alternative = "two.sided") # 잔차에 대한 독립성 검정 수행
```

- $H_0$ : uncorrelated 
- $H_1 : \rho \neq 0$
- 유의확률이 주어진 유의수준보다 크기 때문에 귀무가설을 기각할 수 없다. 
- 즉 잔차들은 독립이라고 할 수 있다.

## `3-(1,a)`

- $H_0$ : $\bf{CompPrice = Income}$
- $H_1$ : $\bf{not}$      $H_0$

```r
library(car)
model = linearHypothesis(full_model, c(0,1,-1,0,0,0,0,0),0)
model
```

```r
# 귀무가설하에서의 모형에서의 SSE
model$RSS[1]
```

$\to$1805.802

$\to$1458.562

```r
model$F[2]
```

$\to$93.32363

## `3-(1,d)`

$\to \bf{TRUE}$

```r
model$`Pr(>F)`[2]
```

$\to$ 6.031627e-20 < 0.05

$\to$ 유의수준 0.05에서 귀무가설 기각 가능
