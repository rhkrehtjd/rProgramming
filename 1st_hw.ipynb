```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
df = pd.read_csv('score.csv')
```

---

```python
df_1 = df.query("gender==1") 
df_1 = df_1.sort_values('written')

df_2 = df.query("gender==2") 
df_2 = df_2.sort_values('written')

plt.plot(range(len(df_1)), df_1.written,label = 'man',color='limegreen' )
plt.plot(range(len(df_2)), df_2.written,label = 'woman',color='red')
plt.axhline(np.mean(df_1.written),color='limegreen'  )
plt.axhline(np.mean(df_2.written),color='red')
plt.legend()
plt.show()
```

```python
df_1 = df.query("gender==1") 
df_1 = df_1.sort_values('practical')

df_2 = df.query("gender==2") 
df_2 = df_2.sort_values('practical')

plt.plot(range(len(df_1)), df_1.practical ,label = 'man',color='limegreen')
plt.plot(range(len(df_2)), df_2.practical ,label = 'woman',color='red')
plt.axhline(np.mean(df_1.practical),color='limegreen'  )
plt.axhline(np.mean(df_2.practical),color='red')
plt.legend()
plt.show()
```
