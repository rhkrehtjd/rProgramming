```r
set.seed(1)

side_num = 6
n_trial = 10^6

dice_result = sample(side_num , n_trial ,replace = TRUE)

index_six = which(dice_result == 6)

index_six[1]
diff(index_six)[1]
mean(c(7,diff(index_six)))
```

```r
num_for_six = c( index_six[1] , diff(index_six))
```

```r
prop_table=prop.table(table(num_for_six))
```

```r
df = data.frame(prop_table)
```

```r
head(df)
```

```r
barplot(df$Freq~df$num_for_six)
```
