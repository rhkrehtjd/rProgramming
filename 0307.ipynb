```r
run.and.value <- function (x) {
  a <- max(rle(x)[[1]])
  b <- rle(x)[[2]][which(rle(x)[[1]] == a)]
  cat("length = ", a, " value = ", b, "\n")
}
```

```r
run.and.value(poisson)
```
