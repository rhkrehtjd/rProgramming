```r

set.seed(1) #   
side = 6     #     
n_trial = 10^6   #   

#  100  ,  
dice_result = sample(side , n_trial ,replace = TRUE)

# 6   
index_six = which(dice_result == 6)

# (1)     6 ?
index_six[1]

# (2)  6   6     
diff(index_six)[1]

# (3) 6      X, X ?
mean(c(index_six[1],diff(index_six))) 
#   diff()   6      index_six[1]    

#  X    

# 6     -> X
num_for_six = c( index_six[1] , diff(index_six))

#  num_for_six  
#   prop.table   
prop_table=prop.table(table(num_for_six))

#    data frame  
df = data.frame(prop_table)

barplot( df$Freq ~ df$num_for_six)

p <- 1/6    # 
x <- seq(1:20) # 20 
Prob <- (1-p)^(x-1) * p # x  

simulation <- df$Freq[1:20]   #  
theory <- (1-p)^(x-1) * p              # x    

mat=rbind(simulation,theory) 


barplot(mat,beside=T,col=c("red","blue"))
legend(20,0.15,c("simulation","theory"),cex=1.3,fill=c("red","blue"))
```
