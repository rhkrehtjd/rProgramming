```r
set.seed(1) 

side_num = 6     # 주사위 총 경우의 수 
n_trial = 10^6   # 주사위 던진 횟수

#
dice_result = sample(side_num , n_trial ,replace = TRUE)

index_six = which(dice_result == 6)

index_six[1]
diff(index_six)[1]
mean(c(index_six[1],diff(index_six)))
```

```r
num_for_six = c( index_six[1] , diff(index_six))
```

```r
prop_table=prop.table(table(num_for_six))
```

```r
df = data.frame(prop_table)
```

```r
head(df)
```

```r
barplot(df$Freq~df$num_for_six)
```

```r
x <- seq(1, 20)

theory <- dexp(x, 1/6)[1:20]
simulation <- df$Freq[1:20]



mat=rbind(theory,simulation)
```

```r
barplot(mat,beside=T)
```
