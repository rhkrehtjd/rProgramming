```r

set.seed(1) # 랜덤 시드 설정
side = 6     # 주사위 총 경우의 수 
n_trial = 10^6   # 주사위 던진 횟수

# 주사위를 100만번 던진 결과, 복원 추출
dice_result = sample(side , n_trial ,replace = TRUE)

# 6의 눈이 나온 경우들
index_six = which(dice_result == 6)

# (1) 주사위를 몇 번째 던져서 6이 나왔는가?
index_six[1]

# (2) 처음 6이 나오고 다시 6이 나올 때까지 주사위를 던진 횟수
diff(index_six)[1]

# (3) 6이 나올 때까지 주사위를 던진 횟수를 X라고하자, X의 평균은?
mean(c(index_six[1],diff(index_six))) 
# 내장 함수 diff() 는 처음 6이 나온 횟수는 계산할 수 없으므로 index_six[1]을 넣어준 후 평균 계산

# 다음은 X의 상대빈도를 막대그래프로 그리는 코드

# 6이 나올 때까지 던진 횟수 -> X
num_for_six = c( index_six[1] , diff(index_six))

# 먼저 num_for_six의 범주별로 카운트
# 그 후 prop.table을 이용하여 상대빈도 계산
prop_table=prop.table(table(num_for_six))

# 상대빈도 계산 결과를 data frame으로 만든 후
df = data.frame(prop_table)

barplot( df$Freq ~ df$num_for_six)

p <- 1/6    # 성공확률
x <- seq(1:20) # 20번 시도
Prob <- (1-p)^(x-1) * p # x번째에 성공할 확률값

simulation <- df$Freq[1:20]   # 모의실험에서의 결과
theory <- (1-p)^(x-1) * p              # x번째에 성공할 확률값의 이론적인 확률

mat=rbind(simulation,theory) 


barplot(mat,beside=T,col=c("red","blue"))
legend(20,0.15,c("simulation","theory"),cex=1.3,fill=c("red","blue"))
```
