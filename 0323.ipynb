```r
sleep <- read.table("http://www.bio.ic.ac.uk/research/mjcraw/therbook/data/sleep.txt", header=TRUE)
```

```r
head(sleep)
```

```r
plot(sleep$Days, sleep$Reaction, type='n', xlab='Days', ylab='Reaction')

s = as.numeric(factor(sleep$Subject))
s

for(k in 1:max(s)){
    x=sleep$Days[s==k]
    y=sleep$Reaction[s==k]
    
    lines(x,y,type="b",col="grey")}
```
