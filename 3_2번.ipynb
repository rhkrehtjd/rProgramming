```r
x = c(.9, 1.3, 2.1, 2.5, 2.4, 1.7, 0.7, 1.2, 1.6)
y = c(2.0, 2.6, 4.3, 5.8, 5.1, 3.2, 1.8, 2.3, 3.0)
```

# 1번

```r
plot(x,y)
abline(lm(y~x))
```

```r
mean_x = mean(x)
mean_y = mean(y)
```

```r
sxy = sum((x-mean_x)*(y-mean_y))
sxx = sum((x-mean_x)^2)
```

```r
b_1 = sxy/sxx
b_0 = mean_y - b_1*mean_x
```

### $\therefore$ 최소제곱법에 의한 회귀직선은 다음과 같다

$y = b_0 + b_1 x$

```r
b_0
b_1
```

### 맞는지 확인해보자

```r
lm(y~x)
```
